# Model Validation Techniques 

```python
# Two Technique 1. collinearity check 2. Residual Analysis
rsq_age=smf.ols('Age~KM+HP+cc+Doors+Gears+QT+Weight',data=toyota4).fit().rsquared
vif_age=1/(1-rsq_age)
rsq_KM=smf.ols('KM~Age+HP+cc+Doors+Gears+QT+Weight',data=toyota4).fit().rsquared
vif_km=1/(1-rsq_KM)
rsq_HP=smf.ols('HP~Age+KM+cc+Doors+Gears+QT+Weight',data=toyota4).fit().rsquared
vif_hp=1/(1-rsq_HP)
rsq_cc=smf.ols('cc~Age+KM+Doors+Gears+QT+Weight+HP',data=toyota4).fit().rsquared
vif_cc=1/(1-rsq_cc)
rsq_Doors=smf.ols('Doors~cc+Age+KM+Gears+QT+Weight+HP',data=toyota4).fit().rsquared
vif_doors=1/(1-rsq_Doors)
rsq_Gears=smf.ols('Gears~Doors+cc+Age+KM+QT+Weight+HP',data=toyota4).fit().rsquared
vif_gears=1/(1-rsq_Gears)
rsq_QT=smf.ols("QT~Gears+Doors+cc+Age+KM+Weight+HP",data=toyota4).fit().rsquared
vif_qt=1/(1-rsq_QT)
rsq_Weight=smf.ols('Weight~Age+KM+HP+Gears+QT+cc+Gears',data=toyota4).fit().rsquared
vif_wt=1/(1-rsq_Weight)

# Making DataFrame
df={'Variable':['Age','KM','HP','cc','Doors','Gears','QT','Weight'],
   'VIF':[vif_age,vif_km,vif_hp,vif_cc,vif_doors,vif_gears,vif_qt,vif_wt]}
vif_df=pd.DataFrame(df)
vif_df
```
